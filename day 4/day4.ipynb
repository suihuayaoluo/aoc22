```python
datalist = []
tem1 = []
tem2 = []
tem3 = []
fir = []
sec = []
thi = []
fou = []

num = 0

f = open("day 4.txt", "r")
for line in f:
    datalist = line.split(",")
    tem1.append(datalist[0])
    tem2.append(datalist[1])
    
for line in tem1:
    a = line.split("-")
    fir.append(a[0])
    thi.append(a[1])
    
    
for line in tem2 :
    b = line.split("-")
    sec.append(b[0])
    tem3.append(b[1])
    
for line in tem3:
    c = line.split()
    fou.append(c[0])
    

for i in range (0, len(fir)):
    if fir[i] <= sec[i] and thi[i] >= fou[i]:
        num += 1
    elif fir[i] >= sec[i] and thi[i] <= fou[i]:
        num += 1

        
print (fir)
print (sec)
print (thi)
print (fou)
```
